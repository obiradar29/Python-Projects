```python
#Arithmetic Game
#Brain Tester
import random
import time
def generate_question():
    op1 = random.randint(1,10)
    op2 = random.randint(1,10)
    operand = random.choice(['+','-','*','//'])
    if operand == '+':
        answer = op1 + op2
    if operand == '-':
        answer = op1 - op2
    if operand == '*':
        answer = op1 * op2
    if operand == '//':
        answer = op1 // op2
    question = str(op1) + operand + str(op2)
    return question,answer
def play_game():
    noofques = int(input("How many questions?"))
    correct = 0
    start_time = time.time()
    for i in range(noofques):
        question,answer = generate_question()
        print(question)
        user_ans = int(input("What is your answer?"))
        if user_ans == answer:
            correct += 1
            print("You are correct")
        else:
            print("You are wrong")
    end_time = time.time()
    print(f"You got {correct} answer correct")
    print("You used {0:.3f} seconds".format(end_time-start_time))
    
play_game()    
```
